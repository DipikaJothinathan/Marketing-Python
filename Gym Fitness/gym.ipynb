```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
workout = pd.read_csv("data/workout.csv")
plt.figure(figsize=(10, 6))
plt.plot(workout['month'], workout['workout_worldwide'], marker='o', linestyle='-', color='b')
plt.title('Worldwide Workout Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Workout Worldwide')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
three_keywords = pd.read_csv("data/three_keywords.csv")
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.plot(three_keywords["month"], three_keywords['home_workout_worldwide'], label='Home Workout Worldwide')
plt.plot(three_keywords["month"], three_keywords['gym_workout_worldwide'], label='Gym Workout Worldwide')
plt.plot(three_keywords["month"], three_keywords['home_gym_worldwide'], label='Home Gym Worldwide')
plt.title('Trends of Workout Keywords Over Time')
plt.xlabel('Month')
plt.ylabel('Search Interest')
plt.legend()
plt.grid(True)
plt.show()
```

```python
workout_geo = pd.read_csv("data/workout_geo.csv")
workout_geo = workout_geo[workout_geo['country'].isin(['United States','Australia','Japan'])]
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.plot(workout_geo["country"], workout_geo['workout_2018_2023'], label='Workout 2018-2023')
plt.title('Workout Trends by Country (2018-2023)')
plt.xlabel('Country')
plt.ylabel('Search Interest')
plt.legend()
plt.grid(True)
plt.show()
```

```python
three_keywords_geo = pd.read_csv("data/three_keywords_geo.csv")
three_keywords_geo = three_keywords_geo[three_keywords_geo['Country'].isin(['Philippines', 'Malaysia'])]
plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
plt.plot(three_keywords_geo["Country"], three_keywords_geo['home_workout_2018_2023'], label='Home Workout 2018-2023')
plt.title('Workout Trends by Country (2018-2023)')
plt.xlabel('Country')
plt.ylabel('Search Interest')
plt.legend()
plt.grid(True)
plt.show()
```
